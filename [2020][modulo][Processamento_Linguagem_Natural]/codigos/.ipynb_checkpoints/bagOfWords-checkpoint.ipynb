# A partir do texto da aula (slide 17)  Construa uma bag of words e identifique o assunto o qual esse texto está relacionado


```python
with open("entrada.txt", "r", encoding="utf-8") as f:
        text = " ".join(f.readlines())
        textoBackup = text
        
```

```python
# print(text)
```

```python
import nltk
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('portuguese')


text = text.lower()

for st in stopwords:
    text = text.replace(' '+st+' ',' ')
#     print(st)
# print("agora sim removemos todas as stopwords")

```

```python

import string
pontosARetirar  = string.punctuation

for p in pontosARetirar:
#         print(p)
        text = text.replace(p,'')
#print("agora sim removi os pontos desnecessários!!!")

#crair um set e colocar cada um dos termos
setTermos = set()
vetorTextoEntrada = text.split(" ")

for termo in vetorTextoEntrada:
#     print(termo)
    setTermos.add(termo)
 
#     setTermos recebe termo

```

```python


#{chave:valor}
#{termo:frequencia}

dictTermos = {}

textoEmLista = textoBackup.split(" ")

# print(textoEmLista)

for termo in setTermos:
    frequencia = textoEmLista.count(termo)
    dictTermos.update({termo:frequencia})

    
dicionarioOrdenado = sorted(dictTermos, key=dictTermos.get, reverse=True)
    
for r in dicionarioOrdenado:
    print(r, dictTermos[r])
    
    
    
```

```python
#agora é a hora de remover os elementos menores e construir a tabela

itemsMaisFrequentes =[]

for item in dicionarioOrdenado:
    if(len(item) > 3) and dictTermos[item] > 2: 
        itemsMaisFrequentes.append(item)
        
print(itemsMaisFrequentes)
```

# Após a construção da bag of words, busque textos na internet e submeta à comparação com a bag of words para identificar se o assunto é relacionado.


```python
with open("novoTexto.txt", "r", encoding="utf-8") as f:
        novoTexto = " ".join(f.readlines())
    
novoTexto = novoTexto.lower()            
novoTextoEmLista = novoTexto.split(" ")
# print(novoTextoEmLista)

listaContagemOcorrencias = []
for itemfrequente in itemsMaisFrequentes:
    listaContagemOcorrencias.append(novoTextoEmLista.count(itemfrequente)) 
    
print(listaContagemOcorrencias)
```

```python
import numpy as np

correspondenciasNovoTexto = listaContagemOcorrencias

import pandas as pd
# pd.columns = itemsMaisFrequentes
ds = pd.DataFrame([correspondenciasNovoTexto])
ds.columns = itemsMaisFrequentes
ds
```
