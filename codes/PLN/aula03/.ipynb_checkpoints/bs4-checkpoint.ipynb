```python
# !pip3 install bs4
from bs4 import BeautifulSoup
# from urllib.request import urlopen
import requests
```

```python
link = 'https://g1.globo.com/loterias/noticia/2020/10/07/mega-sena-concurso-2306-resultado.ghtml'

# html = urlopen(link)
page = requests.get(link)


pagina = BeautifulSoup(page.content,'html')
print(pagina.title)
```

```python
print(pagina.find_all('h1'))
```

```python
pagina.find_all(class_='content-head__subtitle')
```

```python
#coletar todos os links que uma página tem

links  = pagina.find_all('a',href=True)
# links  = pagina.find_all('a')
for link in links:
    print(link.get('href'))
```
