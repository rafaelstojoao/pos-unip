```python
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
plt.figure(figsize=(20,10))
X, y = make_blobs(n_samples=50, n_features=2)
agrupador = KMeans( n_clusters=4,n_init=10)
#n_cluster padr√£o: 8

valoresAgrupados = agrupador.fit(X)
valoresAgrupados = agrupador.fit_predict(X)
instancia_teste = [[3.51520298e+00,  9.49551511e-01]]

agrupador.predict(instancia_teste)
plt.scatter(
    X[valoresAgrupados == 0, 0], X[valoresAgrupados == 0, 1],
    s=50, c='green',
    marker='s', edgecolor='black',
    label='cluster 1'
)

plt.scatter(
    X[valoresAgrupados == 1, 0], X[valoresAgrupados == 1, 1],
    s=50, c='lightblue',
    marker='o', edgecolor='black',
    label='cluster 2'
)

plt.scatter(
    X[valoresAgrupados == 2, 0], X[valoresAgrupados == 2, 1],
    s=50, c='pink',
    marker='v', edgecolor='black',
    label='cluster 3'
)


x = np.random.rand(1)
y = np.random.rand(1)
plt.scatter(x, y, s=200,c='red',marker='o', edgecolor='black', label='')


plt.grid()
plt.show()
```
