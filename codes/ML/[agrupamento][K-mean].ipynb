```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=500, n_features=2)
#X são os valores gerados
#y vetor de cluster de cada valor gerado


# plotagem dos pontos gerados
plt.scatter(X[:, 0], X[:, 1], c='blue', marker='*', edgecolor='black', s=50)
```

```python
from sklearn.cluster import KMeans

agrupador = KMeans( n_clusters=3,n_init=10)
#n_cluster padrão: 8

valoresAgrupados = agrupador.fit(X)
```

```python

    
# plot the centroids
plt.scatter(agrupador.cluster_centers_[:, 0], agrupador.cluster_centers_[:, 1],s=250, marker='.',
    c='red', edgecolor='black',
    label='centróides'
)
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

```python
valoresAgrupados = agrupador.fit_predict(X)
```

```python
instancia_teste = [[3.51520298e+00,  9.49551511e-01]]

agrupador.predict(instancia_teste)
```

```python
plt.scatter(
    X[valoresAgrupados == 0, 0], X[valoresAgrupados == 0, 1],
    s=50, c='green',
    marker='s', edgecolor='black',
    label='cluster 1'
)

plt.scatter(
    X[valoresAgrupados == 1, 0], X[valoresAgrupados == 1, 1],
    s=50, c='lightblue',
    marker='o', edgecolor='black',
    label='cluster 2'
)

plt.scatter(
    X[valoresAgrupados == 2, 0], X[valoresAgrupados == 2, 1],
    s=50, c='pink',
    marker='v', edgecolor='black',
    label='cluster 3'
)

# plot the centroids
plt.scatter(
    agrupador.cluster_centers_[:, 0], agrupador.cluster_centers_[:, 1],s=250, marker='.',
    c='red', edgecolor='black',
    label='centróides'
)
plt.legend(scatterpoints=1)

plt.show()
```
